```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=False)
```

```python
#Monthly Order Trend
monthly_orders = df.groupby('Month')['OrderID'].count()
print(monthly_orders.head())
```

```python
#OUTLIERS IN UNIT PRICE 
Q1_unit = df['UnitPrice'].quantile(0.25)
Q3_unit = df['UnitPrice'].quantile(0.75)
IQR_unit = Q3_unit - Q1_unit
lower_unit = Q1_unit - 1.5 * IQR_unit
upper_unit = Q3_unit + 1.5 * IQR_unit
outliers_unit = df[(df['UnitPrice'] < lower_unit) | (df['UnitPrice'] > upper_unit)]


print("Q1:",round(Q1_unit,2))
print("Q3:",round(Q3_unit,2))
print("IQR",round(IQR_unit,2))
print("Lower Bound:",round(lower_unit,2))
print("Upper Bound:",round(upper_unit, 2))
print("Number of Outliers:",len(outliers_unit))
```

```python
#OUTLIERS IN TOTAL PRICE
Q1_total = df['TotalPrice'].quantile(0.25)
Q3_total = df['TotalPrice'].quantile(0.75)
IQR_total = Q3_total - Q1_total
lower_total = Q1_total - 1.5 * IQR_total
upper_total = Q3_total + 1.5 * IQR_total
outliers_total = df[(df['TotalPrice'] < lower_total) | (df['TotalPrice'] > upper_total)]


print("Q1:",round(Q1_total, 2))
print("Q3:", round(Q3_total, 2))
print("IQR:",round(IQR_total, 2))
print("Lower Bound",round(lower_total, 2))
print("Upper Bound:",round(upper_total, 2))
print("Number of Outliers:", len(outliers_total))
```

```python
#OUTLIERS IN QUANTITY
Q1_qty = df['Quantity'].quantile(0.25)
Q3_qty = df['Quantity'].quantile(0.75)
IQR_qty = Q3_qty - Q1_qty
lower_qty = Q1_qty - 1.5 * IQR_qty
upper_qty = Q3_qty + 1.5 * IQR_qty
outliers_qty = df[(df['Quantity'] < lower_qty) | (df['Quantity'] > upper_qty)]


print("Q1:",round(Q1_qty, 2))
print("Q3:",round(Q3_qty, 2))
print("IQR:", round(IQR_qty, 2))
print("Lower Bound:",round(lower_qty, 2))
print("Upper Bound:", round(upper_qty, 2))
print("Number of Outliers:", len(outliers_qty))

```

```python

#Phase4-THE SO WHAT SUMMARY

# FINDING 1 - BASIC STATISTICS
print(f"Average order value is ${round(df['TotalPrice'].mean(), 2)}")
print(f"Median order value is ${round(df['TotalPrice'].median(), 2)}")
print(f"Highest order value is ${round(df['TotalPrice'].max(), 2)}")
print(f"Lowest order value is ${round(df['TotalPrice'].min(), 2)}")

# FINDING 2 - BEST SELLING PRODUCT
best_product = df['Product'].value_counts().idxmax()
print(f"Most ordered product is: {best_product}")
top_revenue_product = df.groupby('Product')['TotalPrice'].sum().idxmax()
print(f"Product generating most revenue: {top_revenue_product}")

# FINDING 3 - PAYMENT METHOD
top_payment = df['PaymentMethod'].value_counts().idxmax()
print(f"Most used payment method: {top_payment}")
highest_spend_payment = df.groupby('PaymentMethod')['TotalPrice'].mean().idxmax()
print(f"Payment method with highest average spend: {highest_spend_payment}")

# FINDING 4 - ORDER STATUS
top_status = df['OrderStatus'].value_counts().idxmax()
cancelled = df[df['OrderStatus'] == 'Cancelled']['OrderID'].count()
returned = df[df['OrderStatus'] == 'Returned']['OrderID'].count()
delivered = df[df['OrderStatus'] == 'Delivered']['OrderID'].count()
print(f"Most common order status: {top_status}")
print(f"Total Cancelled orders: {cancelled}")
print(f"Total Returned orders: {returned}")
print(f"Total Delivered orders: {delivered}")

# FINDING 5 - REFERRAL SOURCE
top_referral = df['ReferralSource'].value_counts().idxmax()
print(f"Top referral source bringing most customers: {top_referral}")

# FINDING 6 - MONTHLY TREND
df['Month'] = pd.to_datetime(df['Date']).dt.to_period('M')
monthly = df.groupby('Month')['OrderID'].count()
best_month = monthly.idxmax()
worst_month = monthly.idxmin()
print(f"Busiest month: {best_month} with {monthly.max()} orders")
print(f"Quietest month: {worst_month} with {monthly.min()} orders")

# FINDING 7 - OUTLIERS
print(f"UnitPrice outliers found: {len(outliers_unit)}")
print(f"TotalPrice outliers found: {len(outliers_total)}")
print(f"Quantity outliers found: {len(outliers_qty)}")
```
